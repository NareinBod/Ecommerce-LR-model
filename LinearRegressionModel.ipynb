```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
df = pd.read_csv('ecommerce.csv')
df
```

```python
df.info()
#we'll only use columns from 3 to 7
```

```python
plt.scatter(x='Avg. Session Length', y='Yearly Amount Spent', data=df)
plt.title('Avg Session length vs Yearly amount spent')
plt.xlabel('Avg Session length')
plt.ylabel('Yearly Amount spent')
```

```python
plt.scatter(x='Time on App', y='Yearly Amount Spent', data=df)
plt.title('Time on App vs Yearly amount spent')
plt.xlabel('Time on App')
plt.ylabel('Yearly Amount spent')
```

```python
plt.scatter(x='Time on Website', y='Yearly Amount Spent', data=df)
plt.title('Time on Website vs Yearly amount spent')
plt.xlabel('Time on Website')
plt.ylabel('Yearly Amount spent')
```

```python
plt.scatter(x='Length of Membership', y='Yearly Amount Spent', data=df)
plt.title('Length of Membership vs Yearly amount spent')
plt.xlabel('Length of Membership')
plt.ylabel('Yearly Amount spent')
```

```python
df.columns
```

```python
X = df[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership', 'Yearly Amount Spent']]
y = df['Yearly Amount Spent']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state= 42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
#training the model
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
#w values - f(x) = w1x1 + w2x2 + w3x3 ..... wnxn
lm.coef_
```

```python
#After we fit the model we predict now with test values
```

```python
predictions = lm.predict(X_test)
predictions
```

```python
#we plot the data with real values to see how close they are 
plt.scatter(predictions, y_test)
plt.xlabel('predictions')
plt.ylabel('y-test')
plt.title('Evaluation of our linear regression model')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
```

```python
#We are looking at the cost function
print('Mean Absolute error: ', mean_absolute_error(y_test, predictions))
print('Mean Squared error: ', mean_squared_error(y_test, predictions))
print('Root Mean Squared error: ', math.sqrt(mean_absolute_error(y_test, predictions)))
```

```python
#The above data shows the error is very less meaning the model fits very well
```

```python
residuals = y_test - predictions
residuals
```

```python
plt.hist(residuals, bins=30)
plt.show()
```
